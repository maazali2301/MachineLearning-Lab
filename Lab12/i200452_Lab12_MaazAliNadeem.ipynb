# Part 1. For the given dataset (Advertising.csv) implement gradient descent from scratch and find the linear relationship between features and output

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
df=pd.read_csv('Advertising.csv')
df = df.drop("Unnamed: 0", axis=1)
df.head()
```

### 1. Z-Normalization of features and Values

 Z = (value - mean) / std dev

```python
#Features
X = df.drop('sales', axis=1)
X = X.apply(lambda feature:(feature-feature.mean())/feature.std(),axis=0)
X
```

```python
#Values
Y = df['sales']
Y = (Y-Y.mean())/Y.std()
Y = np.array(Y)
Y
```

### 2. Finding Thetas, and Bias Value (Randomly)

```python
bias = random.random()
thetas = [random.random() for i in range(3)]

print("Bias", bias)
print("Thetas", thetas)
```

### 3. Make first prediction

```python
def predict_Yhat(X, thetas, bias):
    return bias + np.dot(X,thetas)
```

```python
Y_hat=predict_Yhat(X,thetas,bias)
```

### 4. Find cost 

```python
def cost_function(Y,Y_hat):
    return np.mean((Y - Y_hat)**2)
```

```python
cost_function(Y,Y_hat)
```

### 5. Update the weights

```python
def update_theta(X,Y,Y_hat,bias,thetas,alpha):
    d_theta0=-2*np.mean(Y-Y_hat)
    d_thetas=(np.dot(-2*(Y-Y_hat),X))/len(Y)
    bias = bias-(alpha*d_theta0)
    thetas = thetas - (alpha*d_thetas)
    return bias, thetas
```

```python
bias,thetas=update_theta(X,Y,Y_hat,bias,thetas,0.01)
print("Bias: ",bias)
print("Thetas: ", thetas)
Y_hat=predict_Yhat(X,thetas,bias)
cost_function(Y,Y_hat)
```

### 6. Make a function for gradient descent

```python
def gradient_descent(X,Y,alpha,i,thetas,bias):
    
    prevthetas=thetas
    prevbias = bias
    prev_cost = 9999999
    
    for x in range(i):
        
        Y_hat=predict_Yhat(X,prevthetas,prevbias)
        bias,thetas=update_theta(X,Y,Y_hat,bias,thetas,alpha)
        cost = cost_function(Y,Y_hat)
        if (cost / prev_cost > 0.99999):
            print(x, ",", cost)
            return prevthetas, prevbias
        prev_cost = cost
        prevthetas = thetas
        prevbias = bias
        print(x, ",", cost)
        
    return prevthetas, prevbias
```

```python
thetas,bias = gradient_descent(X,Y,0.01,1000,thetas,bias)
```

```python
print("Best Theta Values: ", thetas)
print("Best Bias Value: ", bias)
```

## 3. Is linear regression good for classification? If yes, then why, and if no, then why?

Linear Regression is not suitable for classficiation as it aims to fit a best line along the multiple axis of a 
problem with continous data. It will attempt to overfit on a no-linearly dependent set of features. 
In Classification problems, when a new data point occurs, although it may be at a distant end of a classification cluster,
the linear regression models best fit line will stand NULL when a point further from actual line is seen. 

## 4. Why is Mean Squared Error used?

MSE is better than SE and MSA as it caters to normalizing the impact that is created by an outlier value. The error created
by an outlier value is distributed over all over points in the dataset which minimizes its impact. MAE is not as preferred as
MSE, since MAE is not differntiable as it approaches value 0. Absolute of 0 is undefined. 

## 5. Can you find a better estimator than mse?

The R-squared error is an improved estimator when evaluating the loss. 
The R squared error has 2 steps:
    1. Calculate the sum of squared error (same as in MSE)
    2. Calculate the sum of squared error w.r.t data's mean (same as Variance)
R-squared error denotes similarities to the square of Pearson Co-efficeint of Correlation, which is used to identify correlation
magnitudes between two variables. 

Therefore in R-squared, the fitting of line also takes in consideration the correlation of the features and output variable.
