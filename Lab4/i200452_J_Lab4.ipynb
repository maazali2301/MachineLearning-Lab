 Find the probability of the occurrence that 2 people go to the same city

```python
cities = ['A', 'B', 'C', 'D', 'E']
count = 0
for i in cities: 
    for j in cities: 
        for k in cities: 
            if (i==j or j==k or i==k):
                count+=1

print("Probability: ", count/len(cities)**3)
```

Plot the probability distribution.

Categorizing in 3 types
1) All cities are different
2) Two cities are the same
3) All cities are same

```python
cities = ['A', 'B', 'C', 'D', 'E']
c1 = 0
c2 = 0
c3 = 0
for i in cities: 
    for j in cities: 
        for k in cities: 
            if (i==j and j==k):#All three are same
                c3+=1
            if (not i==j and not j==k and not k==i): #None are same
                c1+=1
            elif ((i==j and not k==j) or (j==k and not k==i) or (i==k and not j==i)): #Only two
                c2+=1
                
c1+c2+c3
```

```python
print("No. of Same Cities")
print("X", "|", "Count", "|", "Probability")
print("--------------------------")
print("0", "|", c1, "   |", c1/125 )
print("0", "|", c2, "   |", c2/125 )
print("0", "|", c3, "    |", c3/125 )
```

```python
y = np.array([c1, c2, c3])
labels=["Different Cities", "Any two similar", "All similar"]
explode = [0, 0, 0.2]
plt.pie(y, labels=labels, explode=explode, shadow=True)
plt.show() 
```

#### Is it a discrete distribution or a continuous one?

It is a discrete distribution as the probabilities are assessed on the basis of how many
similar cities are considered in each case. The count of cities can only be discrete with the 
following sample space [0 same cities, 2 people going to same city, and 3 people going to same city]

#### Is it univariate or a multivariate problem?

The problem is univariate as the assessment and statistical analysis is only applied on 
the count of people going to the same city, and no other featured inference is deduced from the problem
